```python
#This code will pick the png images present in the subfolders of the root_path you provided, and
#will paste them on a final image with a white background.You will have 1 final image of each subfolder present in its own folder
```

```python
import math
from PIL import Image
import os

root_path = r"C:\Users\YourUserName\Desktop\ImageFolderName" #or "C:/Users/YourUserName/Desktop/ImageFolderName"

for folder_name in os.listdir(root_path):
    folder_path = os.path.join(root_path, folder_name)

    if not os.path.isdir(folder_path):
        continue

    png_files = [file for file in os.listdir(folder_path) if file.endswith('.png')]

    total_images = len(png_files)
    max_images_per_row = math.ceil(math.sqrt(total_images))

    total_rows = math.ceil(total_images / max_images_per_row)
    total_columns = math.ceil(total_images / total_rows)

    # Calculate the maximum width and maximum height in each row
    max_widths = []
    max_heights = []

    for i in range(total_rows):
        row_files = png_files[i * total_columns: (i + 1) * total_columns]
        row_widths = []
        row_heights = []

        for file in row_files:
            image_path = os.path.join(folder_path, file)
            image = Image.open(image_path).convert('RGBA')  # Convert image to RGBA mode
            width, height = image.size

            row_widths.append(width)
            row_heights.append(height)

            # Set alpha channel to 0 (transparent) for pixels that are black
            for y in range(height):
                for x in range(width):
                    r, g, b, a = image.getpixel((x, y))
                    if r == 0 and g == 0 and b == 0:
                        image.putpixel((x, y), (r, g, b, 0))

            image.save(image_path)  # Save the modified image with transparent background

        max_widths.append(max(row_widths))
        max_heights.append(max(row_heights))

    # Calculate the maximum width and total height
    max_width = max(max_widths)
    total_height = sum(max_heights)

    # Create the final image with the calculated size and white background
    final_image_width = max_width * total_columns
    final_image_height = total_height
    final_image = Image.new('RGB', (final_image_width, final_image_height), (255, 255, 255))  # White background

    # Paste the images onto the final image in the appropriate positions
    current_y = 0

    for i in range(total_rows):
        current_x = 0
        row_files = png_files[i * total_columns: (i + 1) * total_columns]

        for file in row_files:
            image_path = os.path.join(folder_path, file)
            image = Image.open(image_path)

            width, height = image.size
            x_offset = current_x
            y_offset = current_y

            final_image.paste(image, (x_offset, y_offset), image)  # Paste the image with transparency

            current_x += max_width
        current_y += max_heights[i]

    final_image.save(os.path.join(folder_path, "final_image.png"))  # Save the final composite image in the subfolder
```
